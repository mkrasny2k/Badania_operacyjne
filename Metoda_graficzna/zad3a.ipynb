```python
import numpy as np
import matplotlib.pyplot as plt
from gekko import GEKKO

# visualize solution
g = np.linspace(0,10,200)
x,y = np.meshgrid(g,g)
obj = 5*x+y
plt.imshow(((-x+y+5>=0)&(-x+y-5<=0)&(x+2*y>=8)&(2*x+y>=8)&(x>=0)&(y>=0)).astype(int),
    extent=(x.min(),x.max(),y.min(),y.max()),origin='lower',cmap='Greys',alpha=0.3);
# plot constraints
x0 = np.linspace(0, 10, 2000)
y0 = -5+x0   
y1 = 5+x0  
y2 = 4 - 0.5*x0     
#y3 = 8 - 0.5*x0 
y4 = 8 - 2*x0


plt.plot(x0, y0, label=r'$-x1+x2+5\geq0$')
plt.plot(x0, y1, label=r'$-x1+x2-5\leq0$')
plt.plot(x0, y2, label=r'$x1+2x2\geq8$')
#plt.plot(x0, y3, label=r'$-x1-x2+18\geq0$')
plt.plot(x0, y4, label=r'$2x1+x2\geq8$')
#plt.plot([0,0],[0,16], label=r'$x1\geq0$')
CS = plt.contour(x,y,obj,np.arange(-5,27,3))
plt.clabel(CS, inline=1, fontsize=10)
plt.xlim(0,10); plt.ylim(0,10); plt.grid(); plt.tight_layout()
plt.legend(loc=1); plt.xlabel('x1'); plt.ylabel('x2')
plt.savefig('plot.png',dpi=300)
plt.show()
```
